##### I-1 TEMPERATURE MER LANGUE 

```python
import cdsapi
import os

# Créez un dossier pour stocker les fichiers de sortie
output_directory = 'SST_LEF_DATA_DEK'
os.makedirs(output_directory, exist_ok=True)

# Liste des années à récupérer
years = list(range(1961, 2022))  # De 1961 à 2023 inclus

# Créez une instance du client CDS
c = cdsapi.Client(url="https://cds.climate.copernicus.eu/api/v2", key="249316:cfa30517-00b5-4ded-9330-78b7b21349bd")

# Dictionnaire pour les décennies
decade = {"01": range(1, 11), "02": range(11, 21), "03": range(21, 32)}

# Boucle sur les années et télécharge les données
for year in years:
    # Boucle sur les mois
    for month in range(1, 13):
        # Boucle sur les décennies
        for dek, days in decade.items():
            # Définissez le nom du fichier de sortie en utilisant l'année, le mois et la décennie
            output_filename = os.path.join(output_directory, f'sst_lef_{year}_{month}_{dek}.nc')
            
            # Votre requête de récupération de données
            request = {
                'product_type': 'reanalysis',
                'variable': 'sea_surface_temperature',
                'year': str(year),
                'month': str(month).zfill(2),  # Format mois avec 2 chiffres (par exemple, '01' pour janvier)
                'day': [
                    str(day).zfill(2) for day in days  # Format jour avec 2 chiffres
                ],
                'time': [
                    '00:00',"03:00",'06:00',"09:00",'12:00',"15:00",
                    '18:00','21:00',],
                'format': 'netcdf',
                'area': [5, -40, -5, 15],
            }
            
            # Exécutez la requête pour récupérer les données
            c.retrieve('reanalysis-era5-single-levels', request, output_filename)

print("Téléchargement terminé.")
```
