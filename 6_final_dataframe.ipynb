Ce notebook vise à mettre en commun tous les dataframes crées pour constituer le dataset qui va servir pour faire apprendre nos modèles

```python
import pandas as pd
from config import Config
import os 
```

```python
# Définition de la fonction to_numeric_with_nan
def to_numeric_with_nan(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return pd.NA  # Retourne une valeur manquante
```

```python
def label_change(value):
    if pd.isna(value):
        return pd.NA
    elif value == 0.0:
        return '0'
    elif value == 1.0:
        return '1'
    else:
        return value  # Si la valeur est différente de NaN, 0.0 et 1.0, la renvoyer telle quelle
```

### B- DATAFRAMES DES SECHERESSES

#### B-I SPEI DECACE

#### B-I-1 Extreme

```python
sech_ext_1dek=pd.read_csv(str(Config.FILES_TRAITED_PATH) + '/' + str(Config.DATA_1DEK_EXTR))
```

```python
sech_ext_1dek[['Year', 'Month', 'Decade']] = sech_ext_1dek[['Year', 'Month', 'Decade']].applymap(to_numeric_with_nan)
sech_ext_1dek['Label Secheresse'] = sech_ext_1dek['Label Secheresse'].apply(label_change)
sech_ext_1dek.head()
```

##### Merger les dataframes variables decadaires et label secheresse decadaires

```python
data_ext_1dek=variables_dek.merge(sech_ext_1dek, on=['Station', 'Year', 'Month', 'Decade'], how='outer')
data_ext_1dek.head()
#Sauvegarde 
data_ext_1dek.to_csv(str(Config.DATASET_DIR) + '/' + str(Config.DATA_1DEK_EXTR),index=False)
```
